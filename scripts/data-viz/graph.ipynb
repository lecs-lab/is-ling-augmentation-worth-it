```python
import pandas as pd
import numpy as np

df = pd.read_csv("./results-usp-esp.csv")
df["training_size"] = np.where(df["training_size"] == "full", 9575, df["training_size"]).astype(int)
df
```

```python
for column_name in df.columns:
    if column_name.startswith("aug."):
        df[column_name] = np.where(df[column_name] == True, column_name[4:], None)

df
```

```python
df['shortname'] = df.apply(lambda row: x if (x := "+".join((v for k, v in row.dropna().items() if k.startswith("aug")))) != "" else "baseline", axis=1)
df
```

```python
results = df.groupby(["shortname", "training_size"])[["test/BLEU", "test/chrF"]].mean()
results
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 10))

for shortname, group in results.groupby('shortname'):
    group = group.reset_index()
    plt.plot(group['training_size'], group['test/chrF'], marker='o', label=shortname)

# Adding labels and legend
plt.xlabel('log (Training Size)')
plt.xscale('log')
plt.ylabel('Mean test/chrF')
plt.title('Mean test/chrF vs Training Size for Different Shortnames')
plt.legend(title='Shortname')
plt.grid(True)
plt.show()
```
